# Dictionary

Dictionaries are used to store data values in key:value pairs.

```python
sample_dict = {
  "key": "This is a key",
  "value": 1,
  10: 1964
}
print(sample_dict["key"])
print(sample_dict["value"])
print(sample_dict[10])
```

```python
### Example:

input_list = [1, 2, "1", 3, 4, 2, 2, 3, 4, 3, 3]

frequency_dict = {}

for element in input_list:
    if element in frequency_dict:
        frequency_dict[element] += 1
    else:
        frequency_dict[element] = 1

max_element = max(frequency_dict, key=frequency_dict.get)
```

## Functions

### Maximum and minimum



```python

def max2(a,b):
    if a >= b:
        return a
    else:
        return b

def min2(a,b):
    if a <= b:
        return a
    else:
        return b
    
def max3(a,b,c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

def max3v2(a,b,c):
    max_ab = max2(a,b)
    return max2(max_ab,c)

print(max2(5,10))
print(min2(5,10))
print(max3(5,10,10))
print(max3v2(5,10,10))
```
