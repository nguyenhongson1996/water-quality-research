```python
import pandas

csv_file_path = "vietnam_housing_dataset 2.csv"
df = pandas.read_csv(csv_file_path)
df
```

```python
## String split function:
a_string = "This - is - a - string"
string_splited = a_string.split("-")
string_splited
```

```python
## String strip function:
a_string = "   a string   "
a_string.strip()
```

```python
df["Province"] = df["Address"].str.split(",").str[-1].str.strip()
df
```

```python
set(df["Province"].tolist())
```

```python
hanoi_df = df[df["Province"].isin(["HN", "Hà Nội", "Hà Nội", "Hà Nội."])]
hanoi_df
```

```python
hanoi_df = hanoi_df.copy()
hanoi_df["District"] = hanoi_df["Address"].str.split(",").str[-2].str.strip()
hanoi_df
```

```python
set(hanoi_df["District"].tolist())
```

```python
nam_tu_liem_df = hanoi_df[hanoi_df["District"] == "Nam Từ Liêm"].copy()
nam_tu_liem_df
```

```python
smart_city_df = nam_tu_liem_df[nam_tu_liem_df["Address"].str.contains('smart', case=False, na=False)].copy()
smart_city_df
```

## Handle missing value

```python
df_to_process = nam_tu_liem_df[["Area", "Legal status", "Price"]].copy()
df_to_process
```

```python
df_to_process["Legal status"].value_counts().nsmallest().index[0]
```

```python
less_common_value = df_to_process["Legal status"].value_counts().nsmallest().index[0]
df_to_process["Legal status"] = df_to_process["Legal status"].fillna(less_common_value)
df_to_process
```

# Data encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_to_process["actual_label"] = le.fit_transform(df_to_process["Legal status"])
df_to_process
```

```python
mapping = {"Have certificate": 0, "Sale contract": 1}
df_to_process["actual_label"] = df_to_process["Legal status"].map(mapping)
df_to_process
```

# Kmeans

```python
X = df_to_process[["Area", "Price"]]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X)
clusters
```

```python
df_to_process["Cluster"] = clusters
df_to_process
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_to_process["Area"], df_to_process["Price"], 
                     c=df_to_process["Cluster"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Clusters Prediction")
plt.colorbar(scatter, label="Cluster")
```

```python

plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_to_process["Area"], df_to_process["Price"], 
                     c=df_to_process["actual_label"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Actual label")
plt.colorbar(scatter, label="actual_label")
```

# Standardization

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
```

```python
min_max_scaler.fit_transform(df_to_process[["Area", "Price"]])
```

```python
df_to_process[["Area_scaled", "Price_scaled"]] = min_max_scaler.fit_transform(df_to_process[["Area", "Price"]])
df_to_process
```

```python
X = df_to_process[["Area_scaled", "Price_scaled"]]
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X)
df_to_process["New Cluster"] = clusters
```

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_to_process["Area_scaled"], df_to_process["Price_scaled"], 
                     c=df_to_process["New Cluster"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("New Clusters Prediction")
plt.colorbar(scatter, label="New Cluster")
```

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_to_process["Area_scaled"], df_to_process["Price_scaled"], 
                     c=df_to_process["actual_label"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Actual label")
plt.colorbar(scatter, label="actual_label")
```
